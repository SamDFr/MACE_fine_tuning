# Check models and MD input structure

Ce notebook sert à vérifier quels modèles ont été trouvés dans `model/` et quelle structure a été trouvée dans `inputs/md/`.

```python
from pathlib import Path
from ase.io import read
```

```python
ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
MODEL_DIR = ROOT / 'model'
INPUT_MD_DIR = ROOT / 'inputs' / 'md'

model_candidates = []
for pattern in ('*.model', '*.pt', '*.pth'):
    model_candidates.extend(sorted(MODEL_DIR.glob(pattern)))

preferred = ['POSCAR', 'CONTCAR', 'POSCAR-1', 'CONTCAR-1', 'structure.extxyz', 'structure.traj', 'structure.xyz', 'structure.vasp', 'vasprun.xml']
structure_path = None
for name in preferred:
    candidate = INPUT_MD_DIR / name
    if candidate.exists():
        structure_path = candidate
        break

if structure_path is None:
    others = sorted(path for path in INPUT_MD_DIR.iterdir() if path.is_file())
    structure_path = others[0] if others else None

model_candidates, structure_path
```

```python
if not model_candidates:
    raise FileNotFoundError(f'No model found in {MODEL_DIR}')
if structure_path is None:
    raise FileNotFoundError(f'No structure found in {INPUT_MD_DIR}')

atoms = read(str(structure_path), index=-1)
print('Models found:')
for model in model_candidates:
    print('  ', model.name)
print('')
print('Structure:', structure_path)
print('Formula:', atoms.get_chemical_formula())
print('Number of atoms:', len(atoms))
print('Has velocities:', 'momenta' in atoms.arrays)
```
